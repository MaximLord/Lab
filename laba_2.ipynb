```python
#quality - категориальная
#fixed - числовые непрерывные
#volatile - числовые непрерывные
#citric - числовые непрерывные
#residual - числовые непрерывные
#chlorides - числовые непрерывные
#free - числовые непрерывные
#total - числовые непрерывные
#density - числовые непрерывные
#pH - числовые непрерывные
```

```python
def df(column):
 if column['free sulfur dioxide'] < 30:
    val = 'Low'
 elif (column['free sulfur dioxide'] >30) & (column['free sulfur dioxide'] < 70):
    val = 'Middle'
 else :
    val = 'High'
 return val

univ_df['categ_free_sulfur_dioxide'] = univ_df.apply (df, axis=1)
```

```python
univ_df
```

```python
def df(column):
 if column['total sulfur dioxide'] < 30:
    val = 'Low'
 elif (column['total sulfur dioxide'] >30) & (column['total sulfur dioxide'] < 70):
    val = 'Middle'
 else :
    val = 'High'
 return val

univ_df['categ_total_sulfur_dioxide'] = univ_df.apply (df, axis=1)
```

```python
#Построить гистограмму (относительно двух переменных)!!!!!!!!!!!
```

```python
new_dfGist = univ_df[['sulphates','alcohol']].copy ()
sorted_frame = new_dfGist.loc[(new_dfGist['sulphates'] < 500) & (new_dfGist['alcohol'] < 10) & (new_dfGist['sulphates'] > 0)].iloc[1:1000]
figure, ax = plt.subplots()

ax.bar(sorted_frame.sulphates, sorted_frame.alcohol, width=1)
ax.grid(which="major")
ax.minorticks_on()
figure.set_figwidth(10)
figure.set_figheight(6)
```

```python
import seaborn as sns
new_cat_gist = univ_df[['categ_total_sulfur_dioxide', 'sulphates']].copy ()

sorted_frame = sorted_frame.append(new_cat_gist[(new_cat_gist.categ_total_sulfur_dioxide == "Low")].iloc[1:50])
sorted_frame = sorted_frame.append(new_cat_gist[(new_cat_gist.categ_total_sulfur_dioxide == "Middle")].iloc[1:50])
sorted_frame = sorted_frame.append(new_cat_gist[(new_cat_gist.categ_total_sulfur_dioxide == "High")].iloc[1:50])

sns.barplot(
    y="categ_total_sulfur_dioxide",
    x="sulphates",

    order=["Low", "Middle", "High"],
    data=sorted_frame
);
```
